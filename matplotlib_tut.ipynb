```python
import numpy as np
x = np.linspace(0,5,11)     # Get linearly spaced 11 numbers between 0 to 5
```

```python
y = x**2
```

Two ways to represent matplotlib graphs
1. Functional Method
2. Object Oriented method

# Functional Method

```python
plt.plot(x,y, 'r+')
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
# plt.show()       # show() prints the graph and is similar to print('blah blah'), while %matplotlib.inline generates output and is similar to just writing 'blah blah'.
```

```python
# Putting multiple (max 2) plots on same canvas

plt.subplot(1,2,1) #subplot(no_of_rows, no_of_columns, plot_number)
plt.plot(x,y,'ro')

plt.subplot(1,2,2) 
plt.plot(y,x,'bx')

plt.subplot(1,2,1)   # Redraw in plot 1. (There can be at max 2 subplots)
plt.plot(x,x,'g-*')
```

# Object Oriented API method

```python
# Here we instantiate (create) figure objects & then call methods or attributes from that
fig = plt.figure()      # Create a figure object

# Create axis and pass the parameter as list with 4 values [left, bottom, width, height]. These value ranges from 0-1 (basically percentage value)
axes = fig.add_axes([0.2,0.1,0.8,1])  

#Plot the values on axes using the axes object
axes.plot(x,y)

# Set labels using axes object
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
# Plot multiple graphs on same canvas
fig = plt.figure()     # Create a figure object

axes1 = fig.add_axes([0.1,0.1,0.8,0.9])     # Create Main Axes
axes2 = fig.add_axes([0.15,0.52,0.4,0.4])   # Create a graph inside main graph which starts 15% to left of original axes & 52% from the bottom. It occupies 40%x40% of original graph space

axes1.plot(x,y)
axes2.plot(x,y)

axes1.set_title('MAIN GRAPH')
axes2.set_title('SUB GRAPH')
```

```python
# Create multiple figures/ subplots
fig,axes = plt.subplots(nrows=2, ncols=3) # In all there are 2x3=6 figures. This method returns a tuple with array of figs & array of axes

for current_ax in axes:
    current_ax[0].plot(x,y)
    current_ax[1].plot(y,x,'r')
    current_ax[2].plot(x,x,'g')
    
plt.tight_layout()     # This will avoid any overlapping of figs & have them equally spaced.
```

# Figure Size & DPI

```python
fig = plt.figure(figsize=(8,2), dpi=70) # figsize=(width_in_inches, height_in_inches) , dpi=dots/pixels per inch (default =100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2, figsize=(16,2))

axes[0].plot(x,y)
axes[1].plot(y,y)
```

# Legends

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

#axes.plot(x,y,'b')
#axes.plot(y,x,'g')

axes.plot(x,x**2,label="x Square")
axes.plot(x,x**3,label="x Cube")

#axes.legend(loc=10)   #You can Specify location based on location code or string. 
'''
loc = 0 - "best"
loc = 1 - "upper right"
:
loc = 10 - "center"
'''
# You can also specify distance from left & right 0 to 1
axes.legend(loc=(0.1,0.3))
```

# Plot Appearance

Set colors

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

'''
linewidth or lw - specify line width
alpha - transparency (0-1)
linestyle or ls- specify using string or symbols
marker - specify string - '+', 'o' , '*' , ':' , '-' , '--' , ',' , '.', '1', '2', '3', 's', etc.
markersize - specify size of marker'
markeredge - 
markerfacecolor - color inside marker
markeredgecolor - color of marker edge
markeredgewidth - marker edge width
'''

axes.plot(x,x**2, color='magenta', linewidth=5, alpha=0.5)  # Takes names of basic colours
axes.plot(x,x**3, color='#DA4F32',linestyle='steps')  # Use RGB Hex codes.
axes.plot(x,x*10, color='green',linestyle=':', 
          marker='o', markersize='10', markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='purple') 
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, lw=2, ls='--', color='purple')

# Display plot only for x = 0 to 1
axes.set_xlim([0,1])    # [lower_limit, upper_limit]
axes.set_ylim([0,2])    # [lower_limit, upper_limit]
```
